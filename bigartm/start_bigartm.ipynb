```python
files = list(os.walk(LENTAPATH))[0][2]
processed = ['Biznes.csv.bz2']
```

Запускаем оставшиеся файлы на автомате

```python
for file in files:
    if file not in processed:
        df_lenta = pd.read_csv(LENTAPATH+file)
        df_lenta = df_lenta.rename(columns={'1914-lenta/':'date','topic':'predict','topics':'topic',})
        df_lenta = df_lenta[['date','topic','text','predict']]
        df_lenta['text_new'] = df_lenta.text.apply(pr.clean_text)
        df_lenta.text_new = df_lenta.text_new.apply(cn.replaceKeywords)
        df_lenta['lemmatized_text'] = df_lenta.text_new.apply(pr.lemmatization)
        df_lenta.to_csv(LENTAPATH+'out/'+file,index=False,compression='bz2')
        df_interfax = pd.read_csv(INTERFAXPATH+file)
        df_all = pd.concat([df_lenta,df_interfax],axis=0,ignore_index=True)
        print(file ,df_lenta.shape[0]+df_interfax.shape[0]==df_all.shape[0])
        df_all.to_csv('data/all/'+file,index=False,compression='bz2')
        processed.append(file)
        
```

 В итоге мы имее три файла с разными  рубриками с объеденеными датасетами интерфакса и ленты
